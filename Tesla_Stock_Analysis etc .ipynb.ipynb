## Question 2 - Extracting Tesla Revenue Data Using Webscraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "lxml")

tables = soup.find_all("table")
print("Number of tables found:", len(tables))


if len(tables) > 0:
    table = tables[0]
    rows = table.find_all("tr")

    tesla_revenue = []

    for row in rows[1:]:  
        cols = row.find_all("td")
        if len(cols) == 2:
            date = cols[0].text.strip()
            revenue = cols[1].text.strip()
            if revenue != "":
                tesla_revenue.append({"Date": date, "Revenue": revenue})

    tesla_revenue_df = pd.DataFrame(tesla_revenue)

    tesla_revenue_df["Revenue"] = tesla_revenue_df["Revenue"].replace(
        {r"\$": "", ",": ""}, regex=True
    )

    print("\nTesla Revenue Data:")
    print(tesla_revenue_df.head())
    
    tesla_revenue_df.to_csv("tesla_revenue_data.csv", index=False)

```

## Question 5 - Tesla Stock and Revenue Dashboard

```python
import pandas as pd
import plotly.graph_objects as go
import os

df = pd.read_csv("tesla_stock_data.csv", header=[0])  

df.rename(columns=lambda x: x.strip(), inplace=True)
if 'Date' not in df.columns:
    df.reset_index(inplace=True)
    df.rename(columns={'index': 'Date'}, inplace=True)

print("ðŸ§­ Columns in file:", df.columns.tolist())

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
for col in ['Open', 'High', 'Low', 'Close', 'Volume', 'Price']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

df.dropna(subset=['Date'], inplace=True)

fig = go.Figure()

if 'Close' in df.columns:
    fig.add_trace(go.Scatter(x=df['Date'], y=df['Close'], mode='lines', name='Close Price', line=dict(color='cyan')))
elif 'Price' in df.columns:
    fig.add_trace(go.Scatter(x=df['Date'], y=df['Price'], mode='lines', name='Price', line=dict(color='cyan')))
else:
    raise KeyError("No 'Close' or 'Price' column found to plot.")

if 'Volume' in df.columns:
    fig.add_trace(go.Bar(x=df['Date'], y=df['Volume'], name='Volume', yaxis='y2', opacity=0.3))

fig.update_layout(
    title="ðŸš— Tesla Stock Dashboard",
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    yaxis2=dict(title="Volume", overlaying="y", side="right", showgrid=False),
    template="plotly_dark",
    height=600,
)

fig.show()
fig.write_html("tesla_stock_dashboard.html")

print(f"âœ… Dashboard created and saved as 'tesla_stock_dashboard.html' in {os.getcwd()}")
```

## Question 6 - GameStop Stock and Revenue Dashboard

```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import plotly.graph_objects as go

gme_ticker = "GME"
gme_stock_data = yf.download(gme_ticker, start="2020-01-01", end="2025-10-01")
gme_stock_data.reset_index(inplace=True)

print("âœ… GameStop Stock Data Sample:")
print(gme_stock_data.head())

url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
html_data = requests.get(url).text
soup = BeautifulSoup(html_data, "lxml")

tables = soup.find_all("table")
gme_revenue = pd.DataFrame(columns=["Date", "Revenue"])

for table in tables:
    if "GameStop Quarterly Revenue" in str(table):
        rows = table.find_all("tr")
        for row in rows[1:]:
            cols = row.find_all("td")
            if len(cols) == 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip().replace("$", "").replace(",", "")
                gme_revenue = pd.concat(
                    [gme_revenue, pd.DataFrame({"Date": [date], "Revenue": [revenue]})],
                    ignore_index=True
                )

gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"], errors="coerce")
gme_revenue["Date"] = pd.to_datetime(gme_revenue["Date"], errors="coerce")
gme_revenue.dropna(inplace=True)

print("\nâœ… GameStop Revenue Data Sample:")
print(gme_revenue.head())

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=gme_stock_data["Date"], 
    y=gme_stock_data["Close"], 
    mode="lines", 
    name="GameStop Stock Price (USD)", 
    line=dict(color="orange", width=2)
))

fig.add_trace(go.Bar(
    x=gme_revenue["Date"], 
    y=gme_revenue["Revenue"], 
    name="GameStop Revenue (USD Millions)", 
    opacity=0.5,
    yaxis="y2",
    marker_color="lightgreen"
))

fig.update_layout(
    title="ðŸŽ® GameStop Stock & Revenue Dashboard (2020â€“2025)",
    xaxis_title="Date",
    yaxis=dict(title="Stock Price (USD)", side="left"),
    yaxis2=dict(title="Revenue (USD Millions)", overlaying="y", side="right"),
    legend=dict(x=0.02, y=0.95),
    template="plotly_dark",
    height=650
)

fig.show()
fig.write_html("gamestop_dashboard.html")

print("\nâœ… Dashboard created and saved as 'gamestop_dashboard.html'")
```
